<h3 style="color:#FFD700"> GET DISTANCE BETWEEN TWO COORDINATES IN LONGITUDE AND LATITUDE FORMAT </h3>

The function geocode_to_km(fs_lat, fs_long, inc_lat, inc_long) calculates the geographical distance in kilometers between a fire station and an incident location. Both locations are specified by their latitude and longitude coordinates. The function uses the Haversine formula to calculate the great-circle distance between two points on the Earth’s surface, which gives the shortest distance over the Earth’s surface, giving a best approximation of the distance between the points mentioned. The calculated distance is then returned as a float rounded to two decimal places. This function is useful in scenarios where you need to determine the geographical distance between two points when their coordinates are known.

```python
def geocode_to_km(fs_lat, fs_long, inc_lat, inc_long):
    """
    Get distance in kilometers between
    two geographical points that are
    in latitude and longitude format

    Parameters
    -------
    fs_lat : float
        Latitude of firestation

    fs_long : float
        Longitude of firestation

    inc_lat : float
        Latitude of incident area

    inc_long : float
        Longitude of incident area

    Returns
    -------
    Integer
        Distance in kilometers
    """
    earth_radius = 6371  # AVERAGE EARTH RADIUS AT LONDON
    # GEO CODES
    firestation = [fs_lat, fs_long]
    incident = [inc_lat, inc_long]

    # RADIANS CONVERSION
    fs_radians = [radians(geocode) for geocode in firestation]
    inc_radians = [radians(geocode) for geocode in incident]

    # COMPUTE FOR DISTANCE
    distance = haversine_distances([fs_radians, inc_radians])
    distance = distance * earth_radius
    return round(distance[0, 1], 2)
```

<h3 style="color:#FFD700">B.3 Total Reponse Time</h3>

The Total Response Time is the total time it takes from the moment an emergency is reported to the emergency services until the moment the first emergency unit arrives at the scene. This is a derived feature from the summation of Turnout Time, Travel Time, and Attendance Time. Total response time can be further defined through the three aforementioned components:

<h4 style="color:#ADD8E6">B.3.1 Turnout Time</h4>

The Turnout Time measures the time elapsed from when the emergency dispatcher informs the fire unit of an emergency and when the unit leaves the fire station. This is represented by the feature, TurnoutTimeSeconds, and is recorded in seconds.

<h4 style="color:#ADD8E6">B.3.2 Travel Time</h4>

The Travel Time is the time it takes for the fire unit to travel from the fire station to the incident location. Factors such as distance, traffic, and route can impact travel time. This is represented by the feature, TravelTimeSeconds, and is recorded in seconds.

<h4 style="color:#ADD8E6">B.3.3 Attendance Time</h4>

The Attendance Time refers to the arrival of the first fire truck or fire engine at the scene of a fire or other emergency. This is crucial as it marks when firefighting and rescue services begin. Is is measured as the time it takes for the initial responding unit to arrive and start providing firefighting and rescue services. This is represented by the feature, AttendanceTimeSeconds, and is recorded in seconds.

<h3 style="color:#FFD700">C.4 Merging Data</h3>

The team merged the SQL tables of the main datasets - Incidents table and Mobilisation table on the IncidentNumber Field (ID).

We left joined Incident Records (incident_table) into the Mobilisation Records (mobilisation_table) using the IncidentNumber key. We will only be looking at the First mobilisation of pumps through filtering data under PerformanceReports that equal to 1.


<h4 style="color:#ADD8E6">C.4.1 Merge Mobilisation and Incident Table</h4>

```python
with sqlite3.connect('lfb_db.db') as conn:
    conn.execute("""
    CREATE TABLE IF NOT EXISTS incident_merged_mobil_table AS
    SELECT
        IncidentNumber, CalYear, HourOfCall,
        ResourceMobilisationId, Resource_Code,
        PerformanceReporting, DateAndTimeMobilised,
        DateAndTimeMobile, DateAndTimeArrived,
        TurnoutTimeSeconds, TravelTimeSeconds,
        AttendanceTimeSeconds, DateAndTimeLeft,
        DateAndTimeReturned, DeployedFromStation_Code,
        DeployedFromStation_Name, DeployedFromLocation,
        PumpOrder, PlusCode_Code, PlusCode_Description,
        DelayCodeId, DelayCode_Description, DateOfCall,
        TimeOfCall, IncidentGroup, StopCodeDescription,
        SpecialServiceType, PropertyCategory, PropertyType,
        AddressQualifier, Postcode_full, Postcode_district,
        UPRN, USRN, IncGeo_BoroughCode, IncGeo_BoroughName,
        ProperCase, IncGeo_WardCode, IncGeo_WardName,
        IncGeo_WardNameNew, Easting_m,Northing_m, Easting_rounded,
        Northing_rounded, Latitude,Longitude, FRS, IncidentStationGround,
        FirstPumpArriving_AttendanceTime,
        FirstPumpArriving_DeployedFromStation,
        SecondPumpArriving_AttendanceTime,
        SecondPumpArriving_DeployedFromStation,
        NumStationsWithPumpsAttending, NumPumpsAttending,
        PumpCount, PumpHoursRoundUp, [Notional Cost (£)],
        NumCalls
    FROM
        (SELECT
            *
        FROM
            mobilisation_table AS m
        LEFT JOIN
            incidents_table AS i
        ON
            m.IncidentNumber = i.IncidentNumber)
    WHERE
        PerformanceReporting = 1
    """)
```

<h4 style="color:#ADD8E6">C.4.2 Merge Fire Stations Geocodes Table to Merged Mobilisation and Incident Table</h4>

 We link the merged table with geocodes table using DeployedFromStation_Name field. The geocodes table is data that identifies the Longitude and Latitude data of each LFB Fire station.

```python
with sqlite3.connect('lfb_db.db') as conn:
    conn.execute("""
    CREATE TABLE unclean_lfb_table AS
    SELECT
        imm.*,
        StationLatitude,
        StationLongitude
    FROM
        incident_merged_mobil_table AS imm
    INNER JOIN
        fs_geocodes_table as gc
    ON
        imm.DeployedFromStation_Name =
        gc.DeployedFromStation_Name
    """)
```

<h4 style="color:#ADD8E6">C.7.1 Add a New Column: Convert Northing_Rounded and Easting_Rounded to WGS (Latitude, Longitude) </h4>

We used the imported pyproj package here to help calculate and convert Easting and Northing data to Longitude and Latitude to help us explore the data better.

As our analysis of the LFB is centered around responsiveness, measuring distance between the responding fire station and the incident is essential. To do this, we first need the coordinates of the fire station, which has already been completed aboved. We would also need the coordinates of the incident. This is an issue as much of the existing lat-lon coordinates for incidences are missing. It was found that the Northing and Easting coordinate features are the next best substitute as they are present for nearly 100% of records. However, the limitation is that these are rounded and therefore not completely accurate. Additionally, converting the Easting and Northing coordinate features into Lat-Lon coordinates is extremely computationally expensive for this large dataset.

In order to convert Northing and Easting coordinates to Lat-Lon coordinates without exceptionally long-run times, only two points were selected to be directly converted. The average transformation that was performed in these two points was used to approximate the linear transformation to be applied for all points. This reduced the run time from 20 minutes to 0.1 seconds.

```python
with warnings.catch_warnings():
    warnings.filterwarnings("ignore")

    # INTIALIZATION (BNG AND WGS)
    bng = pyproj.Proj(init='epsg:27700')  # British National Grid (OSGB36)
    # World Geodetic System WGS84 (latitude, longitude)
    wgs84 = pyproj.Proj(init='epsg:4326')

    # GET TWO SAMPLE COORDINATES
    easting0 = df_lfb.loc[0, 'Easting_rounded']
    northing0 = df_lfb.loc[0, 'Northing_rounded']
    easting1 = df_lfb.loc[1, 'Easting_rounded']
    northing1 = df_lfb.loc[1, 'Northing_rounded']

    # CONVERT BNG TO WGS
    lon0, lat0 = pyproj.transform(bng, wgs84, easting0, northing0)
    lon1, lat1 = pyproj.transform(bng, wgs84, easting1, northing1)

# GET PARAMETERS FOR CONVERTING NORTHING-EASTING TO LAT-LON
m_lat = (lat1-lat0)/(northing1-northing0)
b_lat = lat0 - m_lat * northing0
m_lon = (lon1-lon0)/(easting1-easting0)
b_lon = lon0 - m_lon * easting0

# USE LINEAR FUNCTION TO CONVERT
df_lfb['IncidentLatitude'] = df_lfb['Northing_rounded'] * m_lat + b_lat
df_lfb['IncidentLongitude'] = df_lfb['Easting_rounded'] * m_lon + b_lon

# PREVIEW
df_lfb[['IncidentLatitude', 'IncidentLongitude',
        'Northing_rounded', 'Easting_rounded']]
```

```python
grouped_df = df_travel.groupby('DeployedFromStation_Name').agg({
    'TravelTimeSeconds': 'sum',
    'DistanceToIncident (KM)': ['sum', 'mean']
})

# Reset column names for the resulting DataFrame
grouped_df.columns = ['Sum_TravelTime', 'Sum_Distance', 'Avg_Distance (KM)']

# Reset the index to make 'DeployedFromStation_Name' a regular column
grouped_df.reset_index(inplace=True)

# Get average speed
grouped_df['Avg_Speed'] = grouped_df['Sum_Distance'] / \
    grouped_df['Sum_TravelTime'] * 3600

# Create Scatter Plot
x = grouped_df['Avg_Distance (KM)']
y = grouped_df['Avg_Speed']

# Create the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, marker='o', s=50, alpha=0.5)

# Set labels and title
plt.xlabel('Average Distance To Incident (KM)')
plt.ylabel('Average Speed (KM/HR)')
plt.title('Average Speed and Distance by Firestation')

# Set axis min and max
max_x = np.ceil(max(x))
plt.xlim(0, max_x)
max_y = np.ceil(max(y) / 10) * 10
plt.ylim(0, max_y)

# Comment here
left_most = grouped_df[grouped_df['Avg_Distance (KM)'] == x.min()]
right_most = grouped_df[grouped_df['Avg_Distance (KM)'] == x.max()]
plumstead = grouped_df[grouped_df['DeployedFromStation_Name'] == 'Plumstead']


plt.annotate(left_most['DeployedFromStation_Name'].values[0],
             (left_most['Avg_Distance (KM)'].iloc[0],
              left_most['Avg_Speed'].iloc[0]),
             textcoords="offset points", xytext=(-10, -10),
             ha='center', fontsize=10, color='red')

plt.annotate(right_most['DeployedFromStation_Name'].values[0],
             (right_most['Avg_Distance (KM)'].iloc[0],
              right_most['Avg_Speed'].iloc[0]),
             textcoords="offset points", xytext=(10, 10),
             ha='center', fontsize=10, color='red')

plt.annotate('Plumstead',
             (plumstead['Avg_Distance (KM)'].iloc[0],
              plumstead['Avg_Speed'].iloc[0]),
             textcoords="offset points", xytext=(10, 10),
             ha='center', fontsize=10, color='red')

# Display the plot
plt.show()

# Linear upward line means that firestations that are expected to face delays
# and traffic are located closer together to reduce travel time
```
